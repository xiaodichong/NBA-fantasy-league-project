```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
player_proj = pd.read_csv(r"C:\Users\jimmy\OneDrive\Desktop\NBA Fantasy league project\NBA data\player_proj_df.csv")
player_info = pd.read_csv(r"C:\Users\jimmy\OneDrive\Desktop\NBA Fantasy league project\NBA data\player_info.csv")
```

```python
player_proj.sample(5)
```

```python
df_real = player_proj.loc[:, ['pts','min','fgm','fga','fg3m','fg3a','ftm','fta','oreb','dreb','ast','stl','tov', 'blk']]
df_proj = player_proj.loc[:, ['proj_pts','proj_min','proj_fgm','proj_fga','proj_fg3m','proj_fg3a','proj_ftm','proj_fta','proj_oreb','proj_dreb','proj_ast','proj_stl','proj_tov', 'proj_blk']]
df_real.sample(5)
```

```python
df_proj.sample(5)
```

```python
se = mean_squared_error(df_real, df_proj)
mse = np.mean(se)
rmse = np.sqrt(mse)
confidence = rmse
```

```python
print('{0} percent confidence in projected {1} per game stats'.format(100 - round(confidence, 2), '2018-19'))
```

```python
se = mean_squared_error(df_real, df_proj)
mse = np.mean(se)
rmse = np.sqrt(mse)
confidence = rmse
print('{0} percent confidence in projected {1} per game stats'.format(100 - round(confidence, 2), '2018-19'))
```

```python
df_real_stats = pd.concat([player_info, df_real], axis=1)
df_real_stats = df_real_stats.dropna()
df_real_stats.sample(5)
```

```python
df_real_stats.rename(columns = 
                    {'proj_season_id' : 'season_id'}, inplace = True)
df_proj_stats = pd.concat([player_info, df_proj], axis=1)
df_proj_stats.rename(columns = {'proj_season_id':'season_id',
                               'proj_pts': 'pts',
                                'proj_min':'min',
                                'proj_fgm':'fgm',
                                'proj_fga':'fga',
                                'proj_fg3m':'fg3m',
                                'proj_fg3a':'fg3a',
                                'proj_ftm':'ftm',
                                'proj_fta':'fta',
                                'proj_oreb':'oreb',
                                'proj_dreb':'dreb',
                                'proj_ast':'ast',
                                'proj_stl':'stl',
                                'proj_tov':'tov',
                               'proj_blk':'blk'}, inplace = True)
```
